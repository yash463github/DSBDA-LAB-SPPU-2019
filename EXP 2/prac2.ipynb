```python
sorted_rscore=sorted(df['reading_score'])
sorted_rscore
```

```python
q1=np.percentile(sorted_rscore,25)
q2=np.percentile(sorted_rscore,75)
print(q1,q2)
```

```python
IQR=q2-q1
```

```python
IQR
```

```python
lwr_bound=q1-(1.5*IQR)

upr_bound = q2+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
r_outliers = []

for i in sorted_rscore: 
    if (i<lwr_bound or i>upr_bound):
        r_outliers.append(i)
        print(r_outliers)
```
